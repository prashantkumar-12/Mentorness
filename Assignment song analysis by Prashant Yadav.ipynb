```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Load the dataset
file_path = 'C:/Users/tulsi/Downloads/songs (5).xlsx'
songs_df = pd.read_excel(file_path)

# Display the first few rows of the dataset
songs_df.head()
```

```python
# Check for missing values
songs_df.isnull().sum()
```

```python
# Handle missing values (example strategy: fill with median or drop)
songs_df = songs_df.dropna(subset=['viewCount', 'likeCount', 'commentCount'])
```

```python
# Convert columns to appropriate data types
songs_df['publishedAt'] = pd.to_datetime(songs_df['publishedAt'])
songs_df['viewCount'] = songs_df['viewCount'].astype(int)
songs_df['likeCount'] = songs_df['likeCount'].astype(int)
songs_df['commentCount'] = songs_df['commentCount'].astype(int)
```

```python
# Display the cleaned dataset
songs_df.info()
```

```python
# Plot distribution of view counts
plt.figure(figsize=(10, 6))
sns.histplot(songs_df['viewCount'], bins=50, kde=True)
plt.title('Distribution of View Counts')
plt.xlabel('View Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of like counts
plt.figure(figsize=(10, 6))
sns.histplot(songs_df['likeCount'], bins=50, kde=True)
plt.title('Distribution of Like Counts')
plt.xlabel('Like Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of comment counts
plt.figure(figsize=(10, 6))
sns.histplot(songs_df['commentCount'], bins=50, kde=True)
plt.title('Distribution of Comment Counts')
plt.xlabel('Comment Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze the distribution of videos across different channels
channel_distribution = songs_df['channelTitle'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=channel_distribution.values, y=channel_distribution.index)
plt.title('Top 10 Channels by Number of Videos')
plt.xlabel('Number of Videos')
plt.ylabel('Channel')
plt.show()
```

```python
# Identify popular tags and their correlation with view counts
tags_exploded = songs_df['tags'].str.split('|').explode()
tags_view_counts = songs_df.explode('tags')[['tags', 'viewCount']]
popular_tags = tags_view_counts.groupby('tags')['viewCount'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=popular_tags.values, y=popular_tags.index)
plt.title('Top 10 Tags by Average View Count')
plt.xlabel('Average View Count')
plt.ylabel('Tag')
plt.show()
```

```python
# Explore how YouTube song video metrics vary over time
songs_df['publish_year'] = songs_df['publishedAt'].dt.year
yearly_views = songs_df.groupby('publish_year')['viewCount'].sum()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_views.index, y=yearly_views.values)
plt.title('Total View Counts Over the Years')
plt.xlabel('Year')
plt.ylabel('Total View Count')
plt.show()
```

```python
# Identify peak publishing times and their impact on engagement
songs_df['publish_hour'] = songs_df['publishedAt'].dt.hour
hourly_engagement = songs_df.groupby('publish_hour')[['viewCount', 'likeCount', 'commentCount']].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_engagement)
plt.title('Average Engagement Metrics by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Count')
plt.legend(['View Count', 'Like Count', 'Comment Count'])
plt.show()
```

```python
# Investigate relationships between likes, comments, and views
plt.figure(figsize=(10, 6))
sns.scatterplot(x='viewCount', y='likeCount', data=songs_df)
plt.title('Relationship Between Views and Likes')
plt.xlabel('View Count')
plt.ylabel('Like Count')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x='viewCount', y='commentCount', data=songs_df)
plt.title('Relationship Between Views and Comments')
plt.xlabel('View Count')
plt.ylabel('Comment Count')
plt.show()
```
